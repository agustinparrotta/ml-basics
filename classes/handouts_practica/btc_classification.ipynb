```python
import ta
import altair as alt
alt.renderers.set_embed_options(theme="dark")
```

```python
dfs = []
for i in range(10):
    dfs.append(
        binance.get_klines("BTCUSDT", "4h", 1000, end_time=None if i==0 else dfs[-1].loc[0, "open_time"])
    )
btc = pd.concat(dfs)
btc.head()
```

```python
btc.shape
```

```python
btc.sort_values("close_time", ascending=True, inplace=True)
btc.reset_index(inplace=True, drop=True)
btc.head()
```

```python
btc.tail()
```

```python
btc_resampled = btc.groupby(btc.index // 6).agg(
    {
        "close_time": [("close_time", "last")],
        "close": [("close", "last")], 
        "high": [("high", "last")],
        "open": [("open", "last")],
        "low": [("low", "last")],
    }
).droplevel(0, axis=1)

btc_resampled.loc[:, "return"] = btc_resampled.loc[:, "close"].pct_change()

btc_resampled.head()
```

```python
c = alt.Chart(btc_resampled.reset_index())

c1 = c.mark_line().encode(
    x = alt.X("close_time:T", title="Date"),
    y = alt.Y("close:Q", title="Price"),
    tooltip = ["open:Q", "low:Q", "high:Q", "close:Q"]
).interactive()

c2 = c.mark_bar().encode(
    x = alt.X("return:Q", bin=alt.Bin(maxbins=60), title="Return"),
    y = alt.Y("count()", title="Count")
).interactive()

c3 = c.mark_rule(color="red").encode(
    x="mean(return):Q",
    size=alt.value(2),
    tooltip=["mean(return):Q"]
)

c1 | c2 + c3
```

```python
btc = ta.add_all_ta_features(
    btc, open="open", high="high", low="low", close="close", volume="volume"
)
```

```python
btc.shape
```

```python
btc.columns
```

```python
cols = ["volume_mfi", "trend_macd", "trend_macd_signal", "momentum_rsi"]
btc.loc[:, cols].isna().sum()
```

```python
def create_target(x:pd.Series, low:float, high:float)->int:
    cum_ret = (1. + x.pct_change().fillna(0.)).cumprod()
    if np.any(cum_ret >= high):
        return 1
    elif np.any(cum_ret <= low):
        return -1
    else:
        return 0
```

```python
btc_transformed = btc.groupby(btc.index//6).agg(
    {
        "close_time": [("close_time", "first")],
        "volume_mfi": [("volume_mfi", "first")],
        "trend_macd": [("trend_macd", "first")],
        "trend_macd_signal": [("trend_macd_signal", "first")],
        "momentum_rsi": [("momentum_rsi", "first")],
        "close": [("target", lambda x: create_target(x, 0.995, 1.005))],
    }
).droplevel(0, axis=1)

btc_transformed.dropna(inplace=True)
btc_transformed.reset_index(drop=True, inplace=True)
btc_transformed.head()
```

```python
btc_transformed.loc[:, "target"].value_counts()
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
train_index = round(0.7*len(btc_transformed))
X_train = btc_transformed.iloc[:train_index].drop(["close_time", "target"], axis=1)
y_train = label_encoder.fit_transform(btc_transformed.iloc[:train_index]["target"])
y_train_cat = to_categorical(y_train, num_classes=3)
```

```python
label_encoder.transform([-1, 0, 1])
```

```python
y_train[:5]
```

```python
y_train_cat[:5]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(
    Dense(8, batch_input_shape=(None, X_train.shape[1]), activation="relu")
)
model.add(
    Dense(
        3, 
        activation="softmax"
    )
)

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(X_train_scaled, y_train_cat, batch_size=16, epochs=200, verbose=1)
```

```python
np.argmax(model.predict(X_train_scaled), axis=1)
```

```python
from sklearn.metrics import classification_report

y_pred = np.argmax(model.predict(X_train_scaled), axis=1)

print(classification_report(y_train, y_pred))
```

```python
btc_train_results = btc_transformed.iloc[:train_index].copy()
btc_train_results.head()
```

```python
btc_train_results.loc[:, ["prob_0", "prob_1", "prob_2"]] = model.predict(X_train_scaled)
btc_train_results.head()
```

```python
btc_train_results.loc[:, "prob_2_bin"] = pd.qcut(btc_train_results.loc[:, "prob_2"], q=10)
btc_train_results.loc[:, "prob_2_bin"] = btc_train_results.loc[:, "prob_2_bin"].astype(str)
btc_train_agg = btc_train_results.groupby("prob_2_bin").agg(
    {
        "target": [("success_rate", lambda x: (x==1).sum()/len(x)), ("count", lambda x: (x==1).sum())]
    }
    ).droplevel(0,axis=1).reset_index()
```
